```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['hg/ha_yield'], bins=30, kde=True)
plt.title("Distribution of Crop Yield")
plt.xlabel("Yield")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Item', y='hg/ha_yield', data=data)
plt.xticks(rotation=90)
plt.title("Average Yield by Crop Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='avg_temp', y='hg/ha_yield', data=data)
plt.title("Temperature vs Crop Yield")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='average_rain_fall_mm_per_year', y='hg/ha_yield', data=data)
plt.title("Rainfall vs Crop Yield")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```
